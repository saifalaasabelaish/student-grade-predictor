```python
import pandas as pd
df=pd.read_csv("../dataset.csv")
```

```python
import pandas as pd

# Assuming df is your DataFrame
# Step 1: Drop all 'Unnamed' columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Step 2: Drop all rows with any NaN values
df = df.dropna()

# Optional: Check the cleaned DataFrame
print(df.head())
```

```python
df
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


# Keep only rows with Grade A-D
df = df[df['GRADE'].isin(['A', 'B', 'C', 'D'])]

# Convert all columns except 'STUDENT ID' and 'GRADE' to category
for col in df.columns:
    if col not in ['STUDENT ID', 'GRADE']:
        df[col] = df[col].astype('category')
```

```python
# First split into train (80%) and temp (20%)
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['GRADE'])

# Then split temp into held-out (10%) and test (10%)
heldout_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42, stratify=temp_df['GRADE'])
```

```python
# Loop through all categorical columns and display categories and their counts
for col in df.columns:
    if df[col].dtype.name == 'category' or df[col].dtype == object:
        print(f"\nColumn: {col}")
        print(df[col].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a dictionary to hold the encoders (for reverse mapping if needed)
label_encoders = {}

# List of columns to encode (exclude STUDENT ID)
columns_to_encode = [col for col in df.columns if col not in ['STUDENT ID']]

# Encode each column
for col in columns_to_encode:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le  # Save encoder for later use

# Show the transformed DataFrame
df.head()
```

```python
# Save preprocessed dataframe to CSV
df.to_csv("../preprocessed_data.csv", index=False)
```
